```python
import numpy as np
import time

```

```python
size=1000_000_000
# python_list
py_list=list(range(size))
start=time.time()
sq_list=(x++2 for x in py_list)
end=time.time()
print(f"python list time= {end-start} seconds")

# numpy array
np_array=np.array(py_list)
start=time.time()

sq_array=np_array **2
end=time.time()
print(f"python array time= {end-start} seconds")
```
