```python
import json
```

```python

def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)

    return data
```

```python
data = load_data("data.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5
    }

    unique_users = set()
    cleaned_data = []

    for user in data:
        # Clean rating - consistent typing
        raw_rating = str(user.get("rating", "")).strip().lower()
        if raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]
        else:
            user["rating"] = raw_rating

        # Handle missing values
        raw_age = user.get("age")
        if raw_age is None:
            user["age"] = None

        # Handle duplicate values - deduplication by name
        name = user.get("name", "").strip()
        if name in unique_users:
            continue

        unique_users.add(name)
        cleaned_data.append(user)

    return cleaned_data
```

```python
data = clean_data(data)
print(data, len(data))
```

```python
def get_insights(data):
    # Average rating
    total_ratings = 0

    for user in data:
        total_ratings += float(user["rating"])

    avg_rating = total_ratings / len(data)
    print(f"Average product rating = {avg_rating}")

    # Percentage of users with poor rating
    poor_ratings = 0

    for user in data:
        if float(user["rating"]) < 3.0:
            poor_ratings += 1

    poor_percentage = (poor_ratings / len(data)) * 100
    print(f"% users with poor rating = {poor_percentage}%")
```

```python
get_insights(data)
```
