```python
import numpy as np
# aggreegation

arr=np.array([1,2,3,4,5])
print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.var(arr))
print(np.count_nonzero(arr))
print(np.std(arr))
print(np.median(arr))

```

```python
arr=np.array([1,2,3,4,5])
print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,3))
```

```python
arr=np.array([1,2,3,4,5])
print(np.log(arr))
print(np.log2(arr))
print(np.exp(arr))
```

```python
print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))


arr=np.array([-1,-2,2,-4,5])

print(np.unique(arr))
print(np.sort(arr))
print(np.abs(arr))


```
