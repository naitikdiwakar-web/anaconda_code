```python
oscar_movies=[
    "THE DARK KNIGHT",
    "the hurt locker",
    "the king's speech",
    "the artist",
    "Argo"
    
]

years=[2008,2009,2010,2011,2012]
oscar_revenue = [1000, 257, 200, 350, 400]

plt.bar(years,oscar_revenue)
plt.title("past data")
plt.xlabel("years")
plt.ylabel("oscar_revenue")


for i in range(len(years)):
    plt.text(
        years[i],
        oscar_revenue[i] + 20,
        str(oscar_revenue[i]),
        ha='center'
    )
plt.ylim(0, max(oscar_revenue)+100)
plt.xlim(min(years) - 1, max(years) +1)

```

```python
import numpy as np
import matplotlib.pyplot as plt

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1000, 257, 200, 350, 400]
non_oscar_revenue = [370, 280, 1005, 78, 345]
third_revenue = [500, 450, 300, 600, 250]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, oscar_revenue, width, label="Oscar Movies")
plt.bar(x, non_oscar_revenue, width, label="Non Oscar Movies")
plt.bar(x + width, third_revenue, width, label="Third Category")

plt.xticks(x, years)

plt.title("Revenue Comparison")
plt.xlabel("Years")
plt.ylabel("Revenue [$M]")
plt.legend()

plt.xticks(x, years)
```

```python

import matplotlib.pyplot as plt
people = ["Alice", "Bob", "Charlie", "David", "Emma",
          "Frank", "Grace", "Henry", "Isabella", "Jack"]

bp = [120, 135, 128, 145, 118,
      140, 125, 150, 122, 138]

age = [25, 42, 35, 50, 22,
       47, 31, 55, 28, 44]


plt.scatter(age, bp, c=bp, s=bp, alpha=0.4, cmap="viridis")

plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("BP")
plt.colorbar(label="BP")
plt.grid(True)





for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i], bp[i]), xytext=(age[i] + 1, bp[i] + 1))

plt.xlim(min(age), max(age) + 10)
plt.ylim(min(bp), max(bp) + 5)

plt.show()
```

```python
cities = [
    "Delhi", "Mumbai", "Kolkata", "Chennai", "Bengaluru",
    "Hyderabad", "Jaipur", "Lucknow", "Pune", "Ahmedabad"
]

winter_temp = [12, 22, 16, 24, 18, 20, 10, 11, 19, 14]  # °C

winter_humidity = [65, 70, 75, 78, 72, 68, 55, 70, 66, 50]  # %

summer_temp = [40, 33, 36 , 38, 30, 37, 42, 41, 34, 43]  # °C

summer_humidity = [35, 80, 75, 70, 60, 45, 25, 40, 55, 20]  # %


plt.scatter(winter_temp,winter_humidity, label="winter")
plt.scatter(summer_temp,summer_humidity,label="summer")

plt.title("Winter vs Summer")
plt.xlabel("Temprature")
plt.ylabel("Humidity")

plt.legend()

```

```python
import matplotlib.pyplot as plt

expenses = ["Rent", "Groceries", "Electricity", "Healthcare", "Dining Out"]
amount = [15000, 5000, 1200, 800, 1500]
colors = ["red", "blue", "green", "orange", "purple"]

explode = [0, 0.1, 0, 0, 0]
plt.pie(amount, 
        labels=expenses,
        autopct="%1.1f%%", 
        colors=colors,
        wedgeprops={"edgecolor": "black", "linewidth": 1,"linestyle": "--"},
        explode=explode,    
        startangle=90,   # rotate chart
        shadow=True 
       )
plt.title("Monthly Expenses Distribution")

plt.style.use("seaborn-v0_8")
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

legit_transactions = [
    125.50, 89.99, 450.00, 32.75, 210.40,
    78.20, 560.10, 99.95, 15.00, 340.60,
    67.80, 1200.00, 45.25, 310.90, 150.75,
    25.99, 480.50, 92.30, 710.20, 55.45,
    180.00, 39.99, 620.15, 140.80, 275.60,
    88.50, 990.00, 135.25, 410.70, 65.35
]

fraud_transactions = [
    500.00, 750.50, 1200.00, 999.99, 1000.75,
    800.20, 2000.00, 1750.40, 1000.60, 950.10,
    2500.00, 3000.50, 1800.80, 2000.25, 3500.90,
    2700.00, 400.75, 3200.10, 1450.55, 5000.00
]

plt.hist(
    legit_transactions,
    bins=20, 
    alpha=0.7,
    color="green",
    edgecolor="blue",
    label="Legit"
)
plt.hist(
    fraud_transactions,
    bins=20, 
    alpha=0.7, 
    color="red",
    edgecolor="black",

    label="Fraud"
)

plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.title("Legit vs Fraud Transactions")
plt.legend()
```
